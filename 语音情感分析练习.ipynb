```python
# -*- coding: utf-8 -*-  
# 读取文件夹地址
import os  
#以递归方式查找所有的音频文件
def listdir(path,list_name):  
#     list_name=list()
    label_list=list()
    for file in os.listdir(path):  
        file_path = os.path.join(path, file)  
        if os.path.isdir(file_path):  
            listdir(file_path, list_name)  
        elif os.path.splitext(file_path)[1]=='.wav':  
            list_name.append(file_path)
```

```python
#读取CASIA数据库
list_name=list()
label_list=listdir('RawData/CASIA database',list_name)
```

```python
label_list=list()
for i in range(len(list_name)):
    name=list_name[i]
    if "angry" in name:
        label_list.append("angry")
#     else:
#         print("nothing")
    if "fear" in name:
        label_list.append("fear")
#     else:
#         print("nothing")
    if "happy" in name:
        label_list.append("happy")
#     else:
#         print("nothing")
    if "neutral" in name:
        label_list.append("neutral")
#     else:
#         print("nothing")
    if "sad" in name:
        label_list.append("sad")
#     else:
#         print("nothing")
    if "surprise" in name:
        label_list.append("surprise")
#     else:
#         print("nothing")
    
```

```python
# 获取所有的标签，这里是识别各种情绪
```

```python
import pandas as pd
```

```python
labels = pd.DataFrame(label_list)
```

```python
labels[:10]
```

```python
# 获取音频特征
```

```python
df = pd.DataFrame(columns=['feature'])
bookmark=0
for index,y in enumerate(list_name):
        #音频导入函数
        # X : 音频的信号值，类型是ndarray,sample_rate : 采样率
        # y 音频路径|sr 采样率（默认22050，但是有重采样的功能）
        # duration 获取音频的时长 |offset 音频读取的时间
        # https://www.cnblogs.com/xingshansi/p/6816308.html
        X, sample_rate = librosa.load(y, res_type='kaiser_fast'
                                      ,duration=2.5,sr=22050*2,offset=0.5)
        #得到采样率
        sample_rate = np.array(sample_rate)
        # MFCC提取：
        # https://www.jianshu.com/p/24044f4c3531
        # 并且求取平均值
        mfccs = np.mean(librosa.feature.mfcc(y=X, 
                                            sr=sample_rate, 
                                            n_mfcc=13),
                        axis=0)
        #获取音频特征
        feature = mfccs
        df.loc[bookmark] = [feature]
        bookmark=bookmark+1  
```

```python
# 将其中的NAN填充成0
rnewdf=rnewdf.fillna(0)
```

```python
# 随机生成数组
newdf1 = np.random.rand(len(rnewdf)) < 0.8
# 获取训练集
train = rnewdf[newdf1]
# 获取测试集
test = rnewdf[~newdf1]
```

```python
train[250:260]
```

```python
# 获取训练特征
trainfeatures = train.iloc[:, :-1]
```

```python
#获取训练标签
trainlabel = train.iloc[:, -1:]
```

```python
#获取测试特征
testfeatures = test.iloc[:, :-1]
```

```python
# 获取测试标签
testlabel = test.iloc[:, -1:]
```

```python

from keras.utils import np_utils

from sklearn.preprocessing import LabelEncoder

X_train = np.array(trainfeatures)

y_train = np.array(trainlabel)

X_test = np.array(testfeatures)

y_test = np.array(testlabel)

lb = LabelEncoder()

y_train = np_utils.to_categorical(lb.fit_transform(y_train))
y_test = np_utils.to_categorical(lb.fit_transform(y_test))
```

```python
y_train
```

```python
X_train.shape
```

```python
#增加一个维度
x_traincnn =np.expand_dims(X_train, axis=2)
# 增加一个维度
x_testcnn= np.expand_dims(X_test, axis=2)
```

```python
# https://blog.csdn.net/qq_30527579/article/details/108853464?utm_medium=distribute.pc_relevant.none-task-blog-BlogCommendFromMachineLearnPai2-1.control&depth_1-utm_source=distribute.pc_relevant.none-task-blog-BlogCommendFromMachineLearnPai2-1.control
```

```python
model = Sequential()

model.add(Conv1D(256, 5,padding='same',
                 input_shape=(216,1)))
model.add(Activation('relu'))
model.add(Conv1D(128, 5,padding='same'))
model.add(Activation('relu'))
model.add(Dropout(0.1))
model.add(MaxPooling1D(pool_size=(8)))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
#model.add(Conv1D(128, 5,padding='same',))
#model.add(Activation('relu'))
#model.add(Conv1D(128, 5,padding='same',))
#model.add(Activation('relu'))
#model.add(Dropout(0.2))
model.add(Conv1D(128, 5,padding='same',))
model.add(Activation('relu'))
model.add(Flatten())
model.add(Dense(6))
model.add(Activation('softmax'))
# 优化器
opt = keras.optimizers.rmsprop(lr=0.00001, decay=1e-6)
```

```python
model.summary()
```

```python
# 开始编译
model.compile(loss='categorical_crossentropy', optimizer=opt,metrics=['accuracy'])
```

```python
cnnhistory=model.fit(x_traincnn
                     , y_train
                     , batch_size=128
                     , epochs=700
                     , validation_data=(x_testcnn, y_test))
```
